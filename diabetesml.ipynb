        #Context

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. 
The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, 
based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection 
of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima 
Indian heritage.

        #Content

The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables 
includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

        #Inspiration

Can you build a machine learning model to accurately predict whether or not the patients in the dataset have diabetes 
or not?

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Importing the dataset
dataset = pd.read_csv('diabetes.csv')
dia = dataset.iloc[:, 0:8].values
target = dataset.iloc[:, 8].values

# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
dia_train, dia_test,target_train, target_test = train_test_split(dia,target,test_size = 0.25,random_state = 0)
```

```python
print(dia_train)
print(dia_test)
print(target_train)
print(target_test)
```

```python
#calling algo of descisiontreeclassifier
clf_dec = DecisionTreeClassifier()
trained_dec = clf_dec.fit(dia_train, target_train)
output_dec = trained_dec.predict(dia_test)
print(output_dec)

#calling algo of KNNclassifier
clf_knn = KNeighborsClassifier(n_neighbors= 7)
trained_knn = clf_knn.fit(dia_train, target_train)
output_knn = trained_knn.predict(dia_test)
print(output_knn)
```

```python
#checking % accuracy for decisiontree
from sklearn.metrics import accuracy_score
checkpct_dec = accuracy_score(target_test, output_dec)
print(checkpct_dec)

#checking % accuracy for KNN
from sklearn.metrics import accuracy_score
checkpct_knn = accuracy_score(target_test, output_knn)
print(checkpct_knn)
```
